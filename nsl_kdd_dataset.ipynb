```python
import pandas as pd
from scipy.io import arff
```

```python
mainpath="../NSL-KDD/"
transformed_datasets_folder = "transformed_datasets/"

file_train_arff="KDDTrain+.arff"
file_test_arff="KDDTest+.arff"

file_complete_traffic = "complete_traffic.csv"
file_anomaly_traffic = "anomaly_traffic.csv"
file_normal_traffic = "normal_traffic.csv"
```

```python
data_train = arff.loadarff(mainpath + file_train_arff)
data_test = arff.loadarff(mainpath + file_test_arff)

df_train = pd.DataFrame(data_train[0])
df_test = pd.DataFrame(data_test[0])
```

```python
df = df_train.append(df_test)
df.head()
```

```python
print(len(df.columns))
print(df.columns)
```

```python
cleanup_labels = {
    "class": {
        b'anomaly': 1, 
        b'normal': 0
        }
    }
df = df.replace(cleanup_labels)
```

```python
for column in df.columns:
        if df[column].dtypes == "object":
            print(column)
            df[[column]] = df[[column]].apply(lambda col:pd.Categorical(col).codes)
```

```python
df.head()
```

```python
label_grouped = df.groupby(['class'])

print(label_grouped['class'].value_counts())

df_anomaly = label_grouped.get_group(1)
df_normal = label_grouped.get_group(0)
```

```python
df_anomaly.head()
```

```python
df_normal.head()
```

```python
df.to_csv(mainpath + transformed_datasets_folder + file_complete_traffic, encoding='utf-8', index=False)
df_anomaly.to_csv(mainpath + transformed_datasets_folder + file_anomaly_traffic, encoding='utf-8', index=False)
df_normal.to_csv(mainpath + transformed_datasets_folder + file_normal_traffic, encoding='utf-8', index=False)
```
